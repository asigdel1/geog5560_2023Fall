Use Python to create maps and perform spatial analysis of housing site for people exiting homelessness in Houston, TX. The spatial analysis will show if the housing units are within a quarter-mile of a public transit stop and if they are in a disadvantaged census tract. 

## Why is this important?
People trying to exit homelessness face an uphill battle even after they find housing. This research project looks at two challanges: lack of private transportation and housing locations in disadvantaged areas. Since many people exiting homelessness do not have access to private transportation, they rely on public transportation to travel to work, medical treatment, recovery programs, and other case management appointments. Without easy access to public transit, people exiting homelessness would struggle to maintain housing. This research project's first objective is to use spatial analysis to determine if single-site housing locations have a public transit stop within a quarter mile distance. 

The second objective is to use spatial analysis to determine if single-site housing locations are in census tracts with a high level of social deprivation. The Social Deprivation Index (SDI) combines several community characteristics at the census tract level that are associated with homelessness, such as limited income, low education, percent of single-family households, and unemployment. The purpose of the second objective is to observe if housing sites are located in census tracts with a high level of SDI, which could lead to re-entry into homelessness.

# Performing data cleaning

```python
def clean_data(result):
    """
    Clean the input DataFrame by replacing missing values and removing rows with incomplete address information.
    Display the count of removed rows and the cleaned DataFrame.

    Parameters:
    - filtered_data (pd.DataFrame): The DataFrame to be cleaned.

    Returns:
    - pd.DataFrame: The cleaned DataFrame.
    """
    try:
        # Count the number of rows before cleaning
        rows_before = len(result)

        # Replace empty strings or other representations of missing values with NaN
        result.replace("", float("nan"), inplace=True)
        result.replace("N/A", float("nan"), inplace=True)  # Add more if needed

        # Remove rows with incomplete address information
        result_clean = result.dropna(subset=["address", "city", "state", "zip"])

        # Count the number of rows before and after cleaning
        rows_after = len(result_clean)
        removed_rows_count = rows_before - rows_after

        # Display the count of removed rows
        print(f"Number of rows removed: {removed_rows_count}")

        # Display the cleaned DataFrame
        print("\nFiltered Data After Removing Rows with Incomplete Address Information:")
        print(result_clean)

        return result_clean

    except Exception as e:
        print(f"Error in clean_and_display_data: {e}")
        return filtered_data  # Return the original DataFrame in case of an error

# Call the function
cleaned_result = clean_data(result)

```

```python
def check_address_format(address_series, address_type):
    """
    Check the formatting of addresses in a given DataFrame column and print the results.

    Parameters:
    - address_series (pd.Series): The pandas Series containing addresses to be checked.
    - address_type (str): A string indicating the type of address (e.g., 'Address', 'City', 'State', 'Zip').

    Returns:
    - None
    """
    print(f"Checking {address_type} formatting:")
    
    for index, address in address_series.items():
        print(f"Index: {index}, {address_type}: {address}")

# Call the function
check_address_format(cleaned_result['address'], 'Address')
check_address_format(cleaned_result['city'], 'City')
check_address_format(cleaned_result['state'], 'State')
check_address_format(cleaned_result['zip'], 'Zip')
```

```python
def clean_addresses(data):
    """
    Clean address-related columns in the input DataFrame.

    Parameters:
    - data (pd.DataFrame): The DataFrame containing address-related columns to be cleaned.

    Returns:
    - pd.DataFrame: The cleaned DataFrame.
    """
    try:
        # Drop rows where the state is not 'TX'
        data = data[data['state'].str.upper() == 'TX']

        # Convert to uppercase and remove leading/trailing whitespaces
        data = data.apply(lambda x: x.str.strip().str.upper() if x.dtype == "object" else x)

        # Remove extra spaces between directional letters and the first letter of the street name
        data['address'] = data['address'].str.replace(r'(\b[NSEW]\.\s*\b)', r'\1', regex=True)

        # Remove periods after directional letters
        data['address'] = data['address'].str.replace(r'(N\.|S\.|E\.|W\.)', r'\1', regex=True)

        # Remove comma at the end of the address
        data['address'] = data['address'].str.rstrip(',')

        # Remove period at the end of the address
        data['address'] = data['address'].str.rstrip('.')

        # Remove specific items in the address1 column
        data['address'] = data['address'].replace({' #6260': '', ', SUITE A': '', 'STE 2': ''}, regex=True)

        # Remove excess spaces
        data = data.apply(lambda x: x.str.replace(r'\s+', ' ', regex=True) if x.dtype == "object" else x)

        # Clean up zip codes: Remove ".0" and keep only the numeric part with exactly five digits
        data['zip'] = data['zip'].astype(str).str.replace(r'\.0$', '', regex=True)
        data = data[data['zip'].str.match(r'^\d{5}$')]

        return data

    except Exception as e:
        print(f"Error in clean_addresses: {e}")
        return data  # Return the original DataFrame in case of an error

# Call the function
final = clean_addresses(result)


```

# Geocoding addresses

```python
geolocator = Nominatim(user_agent="HIC_Locations")

import time

def geocode_address_with_retry(row, max_retries=3):
    """
    Geocode an address using Nominatim with retry mechanism.

    Parameters:
    - row (pd.Series): A row from a DataFrame containing address-related information.
    - max_retries (int): The maximum number of retries in case of geocoding timeouts. Default is 3.

    Returns:
    - shapely.geometry.Point or None: A Point object representing the geocoded location (longitude, latitude)
      or None if geocoding fails after max_retries.
    """
    try:
        # Initialize the geocoder (Nominatim) outside the function
        # geolocator = Nominatim(user_agent="HIC_Locations")  # Uncomment if not initialized elsewhere

        for _ in range(max_retries):
            try:
                # Geocode the address using Nominatim
                location = geolocator.geocode(f"{row['address']}, {row['city']}, {row['state']} {row['zip']}", timeout=10)

                # Return a Point object if location is found, else None
                return Point(location.longitude, location.latitude) if location else None

            except (GeocoderTimedOut, ReadTimeoutError):
                print("Request timed out. Retrying...")
                time.sleep(1)  # Add a short delay before retrying
                continue

        # Return None if geocoding fails after max_retries
        return None

    except Exception as e:
        print(f"Error in geocode_address_with_retry: {e}")
        return None


# Create a new 'geometry' column
final['geometry'] = final.apply(geocode_address_with_retry, axis=1)

# Create a GDF
geo_df = gpd.GeoDataFrame(final, geometry='geometry')

# Plot  
fig, ax = plt.subplots(figsize=(12, 12))
geo_df.plot(ax=ax, alpha=0.5, edgecolor='k', label='Single-Site Housing')

# Add text "Number of points mapped"
num_points_mapped = len(geo_df)
ax.annotate(f"Number of points mapped: {num_points_mapped}", xy=(0.5, -0.1), xycoords="axes fraction", ha="center", fontsize=10)

# Add legend 
legend = ax.legend(loc='upper right', frameon=True)
legend.get_frame().set_edgecolor('black')  # Set edge color for the legend frame

plt.title("Geocoded Single-Site Addresses")
plt.axis("off")

# Save as PNG & display
plt.savefig("geocoded_addresses_map.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
from geopandas.tools import sjoin

# Check CRS
bus_stops_shapefile = bus_stops_shapefile.to_crs(epsg=32615)
filtered_geo_df = filtered_geo_df.to_crs(epsg=32615)

# Do both GeoDataFrames have the same CRS?
bus_stops_shapefile = bus_stops_shapefile.to_crs(filtered_geo_df.crs)

# Change buffer radius from miles to meters
buffer_radius_meters = 0.25 * 1609.34

# Create a GDF for the buffer polygons
buffer_gdf = filtered_geo_df.copy()
buffer_gdf['geometry'] = buffer_gdf['geometry'].buffer(buffer_radius_meters)

# Reset the index
buffer_gdf = buffer_gdf.reset_index(drop=True)

# spatial join
join_result = sjoin(buffer_gdf, bus_stops_shapefile[['geometry']], how='left', predicate='intersects')

# Drop duplicate indices
join_result = join_result.loc[~join_result.index.duplicated(keep='first')]

# Check for transit stops inside the buffer
buffer_gdf['has_transit_stop'] = ~join_result['index_right'].isna()

# Print output
num_housing_sites_with_transit_stop = buffer_gdf['has_transit_stop'].sum()
print(f"Number of housing sites with at least one transit stop within {buffer_radius_meters} miles: {num_housing_sites_with_transit_stop}")
```
